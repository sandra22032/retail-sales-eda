# Item
    The dataset has a Category column. If a row is missing an Item but the Category is "Electronics", by taking the mode value it will fill it with "Item_2_BEV" (which sounds like a Beverage).

```python
# Group by Category and fill Item nulls with the mode of that specific category

data['Item'] = data.groupby('Category')['Item'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# Price Per Unit
    Skewness result was -0.035 (nearly 0). The price distribution is symmetrical and has no extreme outliers

```python
price_avg=data['Price Per Unit'].mean()
price_avg
```

```python
data['Price Per Unit']=data['Price Per Unit'].fillna(price_avg)
```

# Quantity
    The skewness is near zero (the distribution is symmetrical), so we use the Median to ensure we fill nulls with whole numbers for physical items.

```python
qty_med=data['Quantity'].median()
qty_med
```

```python
data['Quantity']=data['Quantity'].fillna(qty_med)
```

# Total Spent

```python
# Logic: Recalculate missing Total Spent values to maintain mathematical consistency (Price * Quantity)

x= data['Total Spent'].isna()
data.loc[x, 'Total Spent'] = data.loc[x, 'Price Per Unit'] * data.loc[x, 'Quantity']
```

# Discount

```python
# Logic: Nulls in discount columns typically mean no discount was applied.

data['Discount Applied'] = data['Discount Applied'].fillna(0)
```

# VISUALIZATION

1. Revenue Contribution by Category :
This vertical bar chart identifies your "Cash Cows." It shows which product categories (e.g., Electronics vs. Food) generate the most money. This is critical for deciding where to allocate marketing budget or which inventory to stock more heavily.

```python
# Grouping by category and summing Total Spent
category_revenue = data.groupby('Category')['Total Spent'].sum().sort_values(ascending=False)

# Vertical Bar Chart
category_revenue.plot(kind='bar', color='seagreen', edgecolor='black', figsize=(10, 6))
plt.title('Total Revenue by Product Category', fontsize=14)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```

2. Monthly Sales Trend (Time Series) :
A line chart of sales over time reveals seasonality (e.g., holiday peaks) and overall business growth. If you see a dip every February or a spike every December, you can prepare your supply chain in advance.

```python
# Converting to datetime if not already done
data['Transaction Date'] = pd.to_datetime(data['Transaction Date'])

# Resampling to Month and summing
monthly_sales = data.set_index('Transaction Date')['Total Spent'].resample('M').sum()

# Line Chart
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='teal')
plt.title('Monthly Sales Trend', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

3. Payment Method Popularity :
Understanding how customers pay (Cash, Credit, Digital Wallet) helps in negotiating transaction fees with banks or choosing which digital wallet integrations to prioritize.

```python
# Counting transactions per payment method
payment_counts = data['Payment Method'].value_counts()

# Set up the figure
plt.figure(figsize=(8, 6))

# Choose a nice color palette
colors = ['#008080', '#20b2aa', '#4db6ac', '#80cbc4']

# Plot a pie chart first
payment_counts.plot(
    kind='pie', 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors, 
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
    pctdistance=0.85
)

# Draw a white circle at the center to transform it into a donut chart
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Market Share of Payment Methods', fontsize=14, pad=20)
plt.ylabel('')  # Clears the default vertical column label
plt.tight_layout()
plt.show()
```

4. Channel Performance: Online vs. In-store :
This compares the "Digital" vs. "Physical" footprint. If Online sales are significantly higher but have a lower average transaction value, it might suggest that customers use the physical store for expensive items and the website for smaller, frequent purchases.

```python
# Grouping by Location
location_sales = data.groupby('Location')['Total Spent'].sum().sort_values(ascending=False)

# Vertical Bar Chart
location_sales.plot(kind='barh', color=['purple', 'brown'], edgecolor='black', figsize=(8, 6.5))
plt.title('Total Sales: Online vs In-store', fontsize=14)
plt.xlabel('Total Revenue ($)', fontsize=12)
plt.ylabel('Location', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

5. Impact of Discounts on Average Spend :
This answers a critical business question: "Do discounts actually make people spend more per transaction?" If the average spend is lower when a discount is applied, your promotions might be eating into your profit without increasing the "basket size."

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# A box plot shows the distribution of 'Total Spent' for both groups
sns.boxplot(
    x='Discount Applied', 
    y='Total Spent', 
    data=data, 
    palette=['#b0bec5', '#008080']
)

# Giving the X-axis categories clean names
plt.xticks(ticks=[0, 1], labels=['No Discount', 'Discount Applied'])

# Styling elements
plt.title('Distribution of Spend: Discount vs No Discount', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Discount Status', fontsize=12)
plt.ylabel('Total Spent ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

# KEY INSIGHTS AND TAKEAWAYS

1. Revenue is remarkably balanced across categories.
All 8 categories sit between roughly $180K–$210K. There's no dominant "cash cow" — Butchers and Electric Household Essentials edge out slightly, but the spread is narrow. Revenue is evenly distributed across categories, suggesting no single product line drives the business — this is either healthy diversification or a missed opportunity to double down on a lead product.

2. Discounts have almost zero effect on basket size.
The chart shows $129.95 (no discount) vs $130.49 (discount applied) — a difference of just $0.54. Discounts are being given away essentially for free with no uplift in spend. That's a significant cost to the business. Discounts may not be working as intended

3. Online vs In-store are nearly equal in revenue.
Both channels are around $750K–$780K. Neither dominates. The near-parity between channels suggests the business has successfully built both digital and physical presence, but there may be untapped opportunity to identify which channel serves which customer segment better.

4. Payment methods are evenly split three ways.
Cash (34.3%), Digital Wallet (33%), Credit Card (32.8%) — almost perfectly equal thirds. No single payment method dominates, which means the business needs to maintain all three infrastructure channels — consolidation could reduce costs.

5. Monthly sales show high volatility with no clear seasonality.
The line chart is spiky throughout 2022–2024 with no consistent December peak or February dip. Sales fluctuate month-to-month without a predictable seasonal pattern, making inventory planning harder. A 3-month rolling average would help smooth this for planning purposes.
